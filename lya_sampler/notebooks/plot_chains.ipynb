## Plot chains 

Notebook to plot chains run sampler object using chainconsumer

```python
import numpy as np
import emcee_sampler
from chainconsumer import ChainConsumer
```

```python
sampler = emcee_sampler.EmceeSampler(read_chain_file=2)        
#sampler.plot_corner(mock_values=True)
paramstrings=[]

for param in sampler.like.free_parameters:
    paramstrings.append(sampler.param_dict[param])
```

```python
chain,lnprob=sampler.get_chain(cube=False)

## Build truth array
truth=[]
for param in sampler.like.free_params:
    truth.append(param.value)
```

```python
c=ChainConsumer()
c.add_chain(chain,parameters=paramstrings)
fig = c.plotter.plot(figsize=(15,15),truth=truth)
fig.show()
```
