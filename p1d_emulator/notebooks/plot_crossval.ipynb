```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_flatten(path):
    with open (path) as json_file:
        list_of_lists=json.load(json_file)
    flat_list=[]
    for sublist in list_of_lists:
        for val in sublist:
            for val2 in val:
                flat_list.append(val2)
    return flat_list
```

```python
theory_flat=load_flatten("/home/chris/Projects/LyaCosmoParams/p1d_emulator/cross_validations/theory_error_full.json")
true_flat=load_flatten("/home/chris/Projects/LyaCosmoParams/p1d_emulator/cross_validations/true_error_full.json")
```

```python
max_error=0.1
plt.figure(figsize=[12,12])
h,xedge,yedge,img=plt.hist2d(true_flat,theory_flat,bins=50,range=[[0,max_error],[0,max_error]])#,cmap=plt.cm.Greys)
plt.xlabel("True error")
plt.ylabel("Theoretical error")
plt.plot(np.linspace(0,max_error,100),np.linspace(0,max_error,100),linestyle="dashed",color="white")
```
