# Compare nuisance parameters in mock data vs default theory 

```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np
## Set default plot size, as normally its a bit too small
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['savefig.dpi'] = 120
mpl.rcParams['figure.dpi'] = 120
import lya_theory
import data_MPGADGET
import camb_cosmo
```

```python
data=data_MPGADGET.P1D_MPGADGET(filename="1024_L90_mimic.json",z_list=[2.0,3.0,4.0])
zs=data.z
```

```python
theory_fid = lya_theory.LyaTheory(zs,emulator=None,verbose=False)
```

```python
theory_mock = lya_theory.get_mock_theory(zs=zs,emulator=None,verbose=False)
```

```python
z_kF=np.linspace(2.0,5.0,100)
kF_kms_fid=theory_fid.kF_model_fid.get_kF_kms(z_kF)
kF_kms_mock=theory_mock.kF_model_fid.get_kF_kms(z_kF)
plt.plot(z_kF,kF_kms_fid,label='fiducial theory')
plt.plot(z_kF,kF_kms_mock,label='mock theory')
plt.xlabel('z')
plt.ylabel('kF [s/km]')
plt.legend()
```
