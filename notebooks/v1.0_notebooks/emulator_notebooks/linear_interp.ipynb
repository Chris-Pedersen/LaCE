# Try linear interpolator from scipy

```python
%matplotlib inline
import numpy as np
import scipy.interpolate
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['savefig.dpi'] = 120
mpl.rcParams['figure.dpi'] = 120
```

```python
# function we want to interpolate
def function(point):
    x=point[0]
    y=point[1]
    f=np.exp(-(x-0.5)**2-(y-0.5)**2)
    #print(point,x,y,f)
    return f
```

```python
# generate random 2D points that will go into the interpolator
ndim=2
ntraining=100
training_points=np.random.rand(ntraining,ndim)
training_values=[function(point) for point in training_points]
```

```python
plt.scatter(training_points[:,0],training_points[:,1],s=10,c=training_values)
plt.xlim([0,1])
plt.ylim([0,1])
plt.colorbar()
```

```python
# setup interpolator
linInter = scipy.interpolate.LinearNDInterpolator(training_points, training_values, fill_value=0.0)
```

```python
# now generate some test points
ntest=10
test_points=np.random.rand(ntest,ndim)
test_values=linInter(test_points)
```

```python
plt.scatter(training_points[:,0],training_points[:,1],s=10,c=training_values,vmin=0.0, vmax=1.0)
plt.scatter(test_points[:,0],test_points[:,1],s=100,c=test_values,vmin=0.0, vmax=1.0)
plt.xlim([0,1])
plt.ylim([0,1])
plt.colorbar()
```
